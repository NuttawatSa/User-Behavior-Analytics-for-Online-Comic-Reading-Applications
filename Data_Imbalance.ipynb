```python
import pandas as pd
from collections import Counter
from imblearn.over_sampling import SMOTE

# โหลดข้อมูล (แก้ path ตามจริง)
df = pd.read_csv("/content/Train_final8_survey_kmitl_encoded(1).csv")

# ระบุ target column
target_col = "Main_apps"
X = df.drop(columns=[target_col])
y = df[target_col]

# ตรวจจำนวนตัวอย่างแต่ละคลาส
print("ก่อนปรับ:", Counter(y))
min_count = min(Counter(y).values())

# กำหนด k_neighbors ให้อัตโนมัติ
k = max(1, min(min_count - 1, 5))   # เลือกไม่เกิน 5 และต้อง < min_count
print("ใช้ k_neighbors =", k)

# ใช้ SMOTE
smote = SMOTE(random_state=42, k_neighbors=k)
X_res, y_res = smote.fit_resample(X, y)

# แสดงผลหลังปรับ
print("หลังปรับ:", Counter(y_res))

# รวมกลับเป็น DataFrame
res_df = pd.DataFrame(X_res, columns=X.columns)
res_df[target_col] = y_res
res_df.to_csv("/content/SMOTE_oversampled.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import TomekLinks

CSV_PATH     = "/content/Train_final8_survey_kmitl_encoded(1).csv"
OUTPUT_PATH  = "/content/SMOTETomek_balanced.csv"
TARGET_CANDS = ["Main_app", "Main_apps"]   # ใส่ชื่อคอลัมน์ที่เป็นไปได้
RANDOM_STATE = 42

df = pd.read_csv(CSV_PATH)

# หา target column
target_col = next((c for c in TARGET_CANDS if c in df.columns), None)
if target_col is None:
    raise ValueError(f"ไม่พบ target column ใน {TARGET_CANDS}")

X = df.drop(columns=[target_col]).copy()
y = df[target_col].copy()

# เข้ารหัสคอลัมน์ที่ไม่ใช่ตัวเลขให้เป็นตัวเลข (กัน error)
for c in X.columns:
    if not np.issubdtype(X[c].dtype, np.number):
        X[c], _ = pd.factorize(X[c], sort=True)

print("ก่อนปรับ:", Counter(y))

# ตั้งค่า k_neighbors สำหรับ SMOTE แบบอัตโนมัติ
min_count = min(Counter(y).values())

k = max(1, min(5, min_count - 1))   # ต้อง < min_count และไม่เกิน 5
smote_tomek = SMOTETomek(
smote=SMOTE(k_neighbors=k, sampling_strategy="not majority", random_state=RANDOM_STATE),
tomek=TomekLinks(),
random_state=RANDOM_STATE
)
X_res, y_res = smote_tomek.fit_resample(X, y)
print("หลัง SMOTE+Tomek:", Counter(y_res))

# รวมกลับเป็น DataFrame
out = pd.DataFrame(X_res, columns=X.columns)
out[target_col] = y_res.values
out.to_csv(OUTPUT_PATH, index=False)
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from imblearn.over_sampling import SMOTE, ADASYN

CSV_PATH     = "/content/Train_final8_survey_kmitl_encoded(1).csv"
OUTPUT_PATH  = "/content/SMOTE_then_ADASYN_partial.csv"
TARGET_CANDS = ["Main_app", "Main_apps"]
RANDOM_STATE = 42

# ปรับสัดส่วนเป้าหมาย (เทียบกับขนาดคลาสใหญ่สุด)
ALPHA1 = 0.6   # เป้าหมายหลัง SMOTE
ALPHA2 = 0.9   # เป้าหมายหลัง ADASYN

# ==== โหลดข้อมูล ====
df = pd.read_csv(CSV_PATH)
target_col = next((c for c in TARGET_CANDS if c in df.columns), None)
if target_col is None:
    raise ValueError(f"ไม่พบ target column ใน {TARGET_CANDS}")

X = df.drop(columns=[target_col]).copy()
y = df[target_col].copy()

# เข้ารหัสคอลัมน์ที่ไม่ใช่ตัวเลขให้เป็นตัวเลข (กัน error ระยะทาง)
for c in X.columns:
    if not np.issubdtype(X[c].dtype, np.number):
        X[c], _ = pd.factorize(X[c], sort=True)

print("[ก่อนปรับ]", Counter(y))
counts = Counter(y)
maj_n = max(counts.values())
min_n = min(counts.values())

k = max(1, min(5, min_n - 1))
print("k_neighbors (auto) =", k)

def partial_strategy(counts, target_n):
    """สร้าง sampling_strategy ให้คลาสที่มีน้อยกว่า target_n ถูกดันขึ้นถึง target_n"""
    return {cls: target_n for cls, n in counts.items() if n < target_n}

# SMOTE
target1 = max(1, int(round(maj_n * ALPHA1)))
strat1 = partial_strategy(counts, target1)
if strat1:
    smote = SMOTE(sampling_strategy=strat1, k_neighbors=k, random_state=RANDOM_STATE)
    X1, y1 = smote.fit_resample(X, y)
else:
    X1, y1 = X, y
print("[หลัง SMOTE]", Counter(y1))

# ADASYN
counts1 = Counter(y1)
target2 = max(1, int(round(maj_n * ALPHA2)))
strat2 = partial_strategy(counts1, target2)
if strat2:
    adasyn = ADASYN(sampling_strategy=strat2, n_neighbors=k, random_state=RANDOM_STATE)
    X2, y2 = adasyn.fit_resample(X1, y1)
else:
    X2, y2 = X1, y1
print("[หลัง ADASYN]", Counter(y2))

# รวมกลับเป็น DataFrame
out = pd.DataFrame(X2, columns=X.columns)
out[target_col] = y2.values
out.to_csv(OUTPUT_PATH, index=False)
```
